## 데이터를 새로 추출할때만 사용
./Data 에 csv파일로 저장

```python
import pandas as pd
import requests
from requests.utils import requote_uri
from time import sleep
```

## League API
challenger 티어에 있는 유저들의 summonerName 을 가져와서 summonerNames에 저장(300명)

```python
URL="https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5"
headers={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36",
    "Accept-Language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
    "Accept-Charset": "application/x-www-form-urlencoded; charset=UTF-8",
    "Origin": "https://developer.riotgames.com",
    "X-Riot-Token": "RGAPI-8c376d7a-d3b2-487b-8fc8-4b2f01eb1507"
}

res=requests.get(URL,headers=headers)

summonerNames=[]
for entry in res.json()["entries"]:
    summonerNames.append(entry["summonerName"])

for name in summonerNames:
    print(name)
```

## Summoner API
summonerNames에 담긴 name으로 유저의 encrypted accountId를 가져와서 accountID에 저장

```python
URL="https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/"


headers= {
    "User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Mobile Safari/537.36",
    "Accept-Language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
    "Accept-Charset": "application/x-www-form-urlencoded; charset=UTF-8",
    "Origin": "https://developer.riotgames.com",
    "X-Riot-Token": "RGAPI-8c376d7a-d3b2-487b-8fc8-4b2f01eb1507"
}

accountID=[]
i=0

for name in summonerNames:
    URL_name=URL+name
    URL_name=requote_uri(URL_name)
    
    res=requests.get(URL_name, headers=headers)
    print(res.headers['X-APP-Rate-Limit-Count'])
    
    # Have to wait because of rate limit
    if("100:120" in res.headers['X-APP-Rate-Limit-Count']):
        print("Request Rate Waiting")
        sleep(120)
        
    try:   
        accountID.append(res.json()['accountId'])
    except:
        accountID.append(-1)
        print(res.status_code)
        print(res.headers)
```

## Match API
accountId로 유저들의 matchID를 가져와서 gameID에 저장

```python
URL = "https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/"
headers={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36",
    "Accept-Language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
    "Accept-Charset": "application/x-www-form-urlencoded; charset=UTF-8",
    "Origin": "https://developer.riotgames.com",
    "X-Riot-Token": "RGAPI-8c376d7a-d3b2-487b-8fc8-4b2f01eb1507"
}

gameID=set()

for ID in accountID:
    # error handling
    if(ID==-1):
        continue
    
    URL_ID=URL+ID
    URL_ID=requote_uri(URL_ID)
    res=requests.get(URL_ID,headers=headers)
    print(res.headers['X-APP-Rate-Limit-Count'])
    # Have to wait because of rate limit
    if("100:120" in res.headers['X-APP-Rate-Limit-Count']):
        print("Request Rate Waiting")
        sleep(120)
    
    if(res.status_code!=200):
        continue
    
    for match in res.json()['matches']:
        matchID=match['gameId']
        #print(matchID,i)
        try:
            gameID.add(matchID)
        except:
            print(res.status_code)
    
```

## Account 정보와 MatchID를 CSV 파일로 저장

```python
from datetime import datetime
date=datetime.today().strftime("_%Y_%m_%d")
```

```python
import csv

gameID_List=list(gameID)
f= open("./Data/matchID"+date+".csv",'w',newline='')
w=csv.writer(f)
w.writerow(["#","matchID"])
for i in range(len(gameID)):
    w.writerow([i+1,gameID_List[i]])
f.close()

f=open("./Data/account"+date+".csv",'w',newline='')
w=csv.writer(f)
w.writerow(["#","summonerName","accountID"])
for i in range(len(summonerNames)):
    w.writerow([i+1,summonerNames[i],accountID[i]])
f.close()
```
