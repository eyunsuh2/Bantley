```python
import pandas as pd
import json
```

```python
df = pd.read_json('./Data/matchData.json')

pd.DataFrame(df)

data=pd.DataFrame()
for i in df['players']:
    L=[]
    for j in i:
        L.append(j['championID'])
    data_temp=pd.DataFrame({'1':[L[0],L[5]],
                '2':[L[1],L[6]],
                '3':[L[2],L[7]],
                '4':[L[3],L[8]],
                '5':[L[4],L[9]]})
    data=pd.concat([data,data_temp],ignore_index=True)
data
```

```python
from roleidentification import pull_data
a=pull_data()
dic={}
min=100
for champID in a.keys():
    pos_list=[]
    for pos in a[champID].keys():
        if(a[champID][pos]>0):
            if(a[champID][pos]<min):
                min=a[champID][pos]
            pos_list.append(pos)
    dic[champID]=pos_list
print(min)
dic
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#correlation heatmap of dataset
def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

correlation_heatmap(data)
```
