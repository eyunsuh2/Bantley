## 챔피언, 챔피언ID 매칭
championName_ID.csv에 저장

```python
import pandas as pd
import json
```

```python
championDict={}
with open('champion.json', 'r',encoding='utf-8') as f:
    data=json.load(f)
    key=data['data'].keys()
    for k in key:
        championDict[(int)(data['data'][k]['key'])]=k
f.close()
championDict
```

```python
import csv
f=open("./Data/championName_ID.csv",'w')
w=csv.writer(f)
w.writerow(["id",'name'])
for key in championDict.keys():
    w.writerow([key,championDict[key]])
w.writerow([-1,'No ban'])
f.close()
print(0)
```
